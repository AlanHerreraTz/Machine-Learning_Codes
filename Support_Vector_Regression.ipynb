#Support Vector Regression (Alan Herrera)

This code serves as an example of how to use Scikit Learn in Python to perform Support Vector Regression (SVR). Support vector machines (SVMs) of the SVR type are employed in regression tasks. It looks for a function that, given an input value, most accurately predicts the continuous output value. Both linear and non-linear kernels can be used with SVR. A non-linear kernel is a more complex function that can identify more subtle patterns in the data than a linear kernel, which is just the dot product of two input vectors. The characteristics of the data and the difficulty of the task determine which kernel to use.
Three distinct kernels—RBF, linear, and polynomial—are used by the code to fit an SVR model and generate sample data. Plotting the expected output for each kernel as a line with a distinct color corresponds to the input data, which is represented as a scatter plot with orange dots. The labels "data" and "target" are placed on the x- and y-axes, respectively. The plot is called "Support Vector Regression." Which line goes with which kernel is indicated by the legend.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVR

X = np.sort(5 * np.random.rand(40, 1), axis=0)
y = np.sin(X).ravel()
y[::5] += 3 * (0.5 - np.random.rand(8))

svr_rbf = SVR(kernel="rbf", C=100, gamma=0.1, epsilon=.1)
svr_lin = SVR(kernel="linear", C=100, gamma="auto")
svr_poly = SVR(kernel="poly", C=100, gamma="auto", degree=3, epsilon=.1, coef0=1)

lw = 2
plt.scatter(X, y, color='darkorange', label='data')
plt.plot(X, svr_rbf.fit(X, y).predict(X), color='navy', lw=lw, label='RBF model')
plt.plot(X, svr_lin.fit(X, y).predict(X), color='c', lw=lw, label='Linear model')
plt.plot(X, svr_poly.fit(X, y).predict(X), color='cornflowerblue', lw=lw, label='Polynomial model')
plt.xlabel('data')
plt.ylabel('target')
plt.title('Support Vector Regression')
plt.legend()
plt.show()
```

#Explanation of the code:

The Python code uses Scikit Learn to create sample data and fit a Support Vector Regression (SVR) model. The code is explained in full below:

* import numpy as np: Import the NumPy library as np.

* y = np.sin(X).ravel(): Compute the sine of each element in X and flatten the resulting array into a 1D array. This creates a 40x1 array of target values.

* y[::5] += 3 * (0.5 - np.random.rand(8)): Add some random noise to every 5th element in y.

* svr_rbf = SVR(kernel="rbf", C=100, gamma=0.1, epsilon=.1): Create an instance of the SVR class with a radial basis function (RBF) kernel, regularization parameter C set to 100, kernel coefficient gamma set to 0.1, and epsilon-insensitive loss parameter epsilon set to 0.1.

* svr_lin = SVR(kernel="linear", C=100, gamma="auto"): Create an instance of the SVR class with a linear kernel, regularization parameter C set to 100, and kernel coefficient gamma set to “auto”.

* svr_poly = SVR(kernel="poly", C=100, gamma="auto", degree=3, epsilon=.1, coef0=1): Create an instance of the SVR class with a polynomial kernel of degree 3, regularization parameter C set to 100, kernel coefficient gamma set to “auto”, epsilon-insensitive loss parameter epsilon set to 0.1, and independent term in kernel function coef0 set to 1.

* lw = 2: Set the line width for the plots to 2.

* plt.scatter(X, y, color='darkorange', label='data'): Plot the input data as a scatter plot with orange dots.

* plt.plot(X, svr_rbf.fit(X, y).predict(X), color='navy', lw=lw, label='RBF model'): Fit the SVR model with RBF kernel to the input data and plot the predicted output as a blue line.

* plt.plot(X, svr_lin.fit(X, y).predict(X), color='c', lw=lw, label='Linear model'): Fit the SVR model with linear kernel to the input data and plot the predicted output as a cyan line.

* plt.plot(X, svr_poly.fit(X, y).predict(X), color='cornflowerblue', lw=lw, label='Polynomial model'): Fit the SVR model with polynomial kernel to the input data and plot the predicted output as a light blue line.

* plt.xlabel('data'): Label the x-axis as “data”.

* plt.ylabel('target'): Label the y-axis as "target
